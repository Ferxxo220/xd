## Carga de Datos

```
import pandas as pd
import numpy as np
df = pd.read_csv('PlayTennis.csv')
```

## Limpieza Inicial
Se eliminaron columnas irrelevantes: ['Unnamed: 0']

```
# Eliminar columnas índice basura
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

## Limpieza de Nulos

```
# Imputación de nulos
df.fillna(method='ffill', inplace=True)
```

## Encoding Automático

```
# Encoding: Texto a Números
for c in df.select_dtypes('object').columns:
    df[c] = df[c].astype('category').cat.codes
```

## Exportar CSV Limpio

```
df.to_csv('PlayTennis_clean_encoded.csv', index=False)
```
